```python
import numpy as np
import matplotlib.pyplot as plt

def plot_quadratic_equation(a, b, c):
    x = np.linspace(-10, 10, 400)
    y = a * x**2 + b * x + c
    
    plt.plot(x, y)
    plt.axhline(0, color='black', lw=0.5)
    plt.axvline(0, color='black', lw=0.5)
    
    discriminant = b**2 - 4*a*c
    if discriminant > 0:
        # 有兩個實根
        root1 = (-b + np.sqrt(discriminant)) / (2*a)
        root2 = (-b - np.sqrt(discriminant)) / (2*a)
        plt.plot(root1, 0, 'ro', label='Root')
        plt.plot(root2, 0, 'ro')
        plt.annotate(f'({root1:.2f}, 0)', (root1, 0), textcoords="offset points", xytext=(0,10), ha='center')
        plt.annotate(f'({root2:.2f}, 0)', (root2, 0), textcoords="offset points", xytext=(0,10), ha='center')
    elif discriminant == 0:
        # 有一個實根（重根）
        root = -b / (2*a)
        plt.plot(root, 0, 'ro', label='Root')
        plt.annotate(f'({root:.2f}, 0)', (root, 0), textcoords="offset points", xytext=(0,10), ha='center')
    
    vertex_x = -b / (2*a)
    vertex_y = a * vertex_x**2 + b * vertex_x + c
    plt.plot(vertex_x, vertex_y, 'go', label='Vertex')
    plt.annotate(f'({vertex_x:.2f}, {vertex_y:.2f})', (vertex_x, vertex_y), textcoords="offset points", xytext=(0,10), ha='center')
    
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(f'Quadratic Equation: Y = {a}x^2 + {b}x + {c}')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
# 測試用例
plot_quadratic_equation(-3, 4, 7)
plot_quadratic_equation(5, -4, 9)
```
